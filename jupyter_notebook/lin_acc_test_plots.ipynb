```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20,15]

data_dir = "/data1/kranti/pratyush/quadrotor_model/data_high_rates/acc/"
pred_dir = "/data1/kranti/pratyush/quadrotor_model/s_vel/predictions/"
normalize_dir = "/data1/kranti/pratyush/quadrotor_model/s_vel/"
```

```python
circle_gt = np.load(data_dir + 'a_circle.npy')
circle_nn = np.load(pred_dir + 'circle_prediction.npy')

test_traj_1_gt = np.load(data_dir + 'a_test_traj_1.npy')
test_traj_1_nn = np.load(pred_dir + 'test_traj_1_prediction.npy') 


y_sin_z_gt = np.load(data_dir + 'a_y_sin_z.npy')
y_sin_z_nn = np.load(pred_dir + 'y_sin_z_prediction.npy')
```

```python
min_val_Y = np.load(normalize_dir + 'min_val_Y.npy')
max_val_Y = np.load(normalize_dir + 'max_val_Y.npy')

for col in range(0,circle_gt.shape[1]):
    circle_nn[:,col] = (( (circle_nn[:,col]) * ( max_val_Y[col] - min_val_Y[col] ) ) + min_val_Y[col] )

    
for col in range(0,test_traj_1_gt.shape[1]):
    test_traj_1_nn[:,col] = (( (test_traj_1_nn[:,col]) * ( max_val_Y[col] - min_val_Y[col] ) ) + min_val_Y[col] )

for col in range(0,y_sin_z_gt.shape[1]):
    y_sin_z_nn[:,col] = (( (y_sin_z_nn[:,col]) * ( max_val_Y[col] - min_val_Y[col] ) ) + min_val_Y[col] )
```

### Test trajectory : Circle

```python
fig1, ax1 = plt.subplots(nrows=3,ncols=1)

ax1[0].set_title('acc :x-direction')
ax1[0].plot(circle_gt[0:3000,0],'r',label='NN_prediction' )
ax1[0].plot(circle_nn[0:3000,0],'b',label='measured')
ax1[0].legend()
ax1[0].set_xlabel('time')
ax1[0].set_ylabel(r'$m/s^2$')

ax1[1].set_title('acc :y-direction')
ax1[1].plot(circle_gt[0:3000,1],'b' ,label='measured')
ax1[1].plot(circle_nn[0:3000,1],'r' ,label='NN_prediction' )
ax1[1].legend()
ax1[1].set_xlabel('time')
ax1[1].set_ylabel(r'$m/s^2$')

ax1[2].set_title('acc :z-direction')
ax1[2].plot(circle_gt[0:3000,2],'b',label='measured')
ax1[2].plot(circle_nn[0:3000,2],'r' ,label='NN_prediction' )
ax1[2].legend()
ax1[2].set_xlabel('time')
ax1[2].set_ylabel(r'$m/s^2$')

plt.show()
```

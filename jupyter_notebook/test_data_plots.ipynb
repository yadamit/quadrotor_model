```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20,5]
```

```python
data_dir      = "/data1/kranti/pratyush/quadrotor_model/s_omega/data_high_rates/nn2/"
prediction    = "/data1/kranti/pratyush/quadrotor_model/s_omega/predictions/"
normalize_dir = "/data1/kranti/pratyush/quadrotor_model/s_omega/"

circle_gt = np.load(data_dir + 'nn2_circle_grd_truth.npy')
circle_nn = np.load(prediction + 'circle_prediction.npy')

test_traj_1_gt = np.load(data_dir + 'nn2_test_traj_1_grd_truth.npy')
test_traj_1_nn = np.load(prediction + 'test_traj_1_prediction.npy') 
```

```python
min_val_Y = np.load(normalize_dir + 'min_val_Y.npy')
max_val_Y = np.load(normalize_dir + 'max_val_Y.npy')

for col in range(0,circle_gt.shape[1]):
    circle_nn[:,col] = (( (circle_nn[:,col]) * ( max_val_Y[col] - min_val_Y[col] ) ) + min_val_Y[col] )

    
for col in range(0,test_traj_1_gt.shape[1]):
    test_traj_1_nn[:,col] = (( (test_traj_1_nn[:,col]) * ( max_val_Y[col] - min_val_Y[col] ) ) + min_val_Y[col] )
```

```python
fig, ax = plt.subplots(nrows=1,ncols=3)
ax[0].plot(circle_nn[:,0],'r' )
ax[0].plot(circle_gt[:,0],'b' )

ax[1].plot(circle_gt[:,1],'b' )
ax[1].plot(circle_nn[:,1],'r' )

ax[2].plot(circle_gt[:,2],'b' )
ax[2].plot(circle_nn[:,2],'r' )

plt.show()
```

```python
fig1, ax1 = plt.subplots(nrows=1,ncols=3)
ax1[0].plot(test_traj_1_gt[:,0],'r' )
ax1[0].plot(test_traj_1_nn[:,0],'b' )

ax1[1].plot(test_traj_1_gt[:,1],'b' )
ax1[1].plot(test_traj_1_nn[:,1],'r' )

ax1[2].plot(test_traj_1_gt[:,2],'b' )
ax1[2].plot(test_traj_1_nn[:,2],'r' )

plt.show()
```

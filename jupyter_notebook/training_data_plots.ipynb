```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 15]
```

```python
data_dir = "/data1/kranti/pratyush/quadrotor_model/s_omega/data_high_rates/nn2/"
# data_dir    = "/data1/kranti/pratyush/quadrotor_model/s_omega/data/nn2/"


rand_x_y = np.load(data_dir + 'nn2_rand_x_y_grd_truth.npy')
x_rand_y = np.load(data_dir + 'nn2_x_rand_y_grd_truth.npy')

rand_neg_x_y = np.load(data_dir + 'nn2_rand_x_neg_y_grd_truth.npy')
rand_x_neg_y = np.load(data_dir + 'nn2_rand_x_neg_y_grd_truth.npy')

x_sin_y = np.load(data_dir + 'nn2_x_sin_y_grd_truth.npy')
```

```python
fig,ax = plt.subplots(nrows=3, ncols=3)
ax[0][0].set_title('rand_x_y : roll alpha')
ax[0][0].plot(rand_x_y[0:1000,0],'b')


ax[0][1].set_title('rand_x_y : pitch alpha')
ax[0][1].plot(rand_x_y[0:1000,1],'g')

ax[0][2].set_title('rand_x_y : yaw alpha')
ax[0][2].plot(rand_x_y[0:1000,2],'r')


ax[1][0].set_title('x_rand_y : roll alpha')
ax[1][0].plot(x_rand_y[0:1000,0],'b')

ax[1][1].set_title('x_rand_y : pitch alpha')
ax[1][1].plot(x_rand_y[0:1000,1],'g')

ax[1][2].set_title('x_rand_y : yaw alpha')
ax[1][2].plot(x_rand_y[0:1000,2],'r')


ax[2][0].set_title('x_sin_y : roll alpha')
ax[2][0].plot(x_sin_y[0:1000,0],'b')

ax[2][1].set_title('x_sin_y : pitch alpha')
ax[2][1].plot(x_sin_y[0:1000,1],'g')

ax[2][2].set_title('x_sin_y : yaw alpha')
ax[2][2].plot(x_sin_y[0:1000,2],'r')

plt.show()
```
